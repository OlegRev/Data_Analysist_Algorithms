Реализовать оптимизацию методом градиентного спуска для $x^2(np.sin(0,5x)^2+1)$. <br>
Какие параметры шага и количества итераций оптимальны?<br>
При каких значениях шага оптимизиция не выходит из локальных минимумов?


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50, 50, 5000)
def func2(_x): return (_x**2)*(np.sin(0.5*_x)**2+1)
def grad2(_x): return 2*(_x)+2*_x*np.sin(0.5*_x)**2+_x**2*np.sin(0.5*_x)*np.cos(0.5*_x)
def calculate_grad(func, x0, delta=0.01): return (func(x0+delta)-func(x0))/delta
def calculate_grad2(func, x0, delta=0.01): return (func(x0+delta)-func(x0-delta))/(2*delta)
```

```python
x0=np.random.uniform(-50, 50)
print('true grad', grad2(x0))
print('method1', calculate_grad(func2, x0))
print('method2', calculate_grad2(func2, x0))
```

```python
calculate_grad2(func2,x0,delta=0.019)
```

```python
for delta_x in np.linspace(0, 1, 100000):
    if round(calculate_grad2(func2,x0,delta=delta_x), ndigits=8)==round(grad2(x0),ndigits=8):
        print(f'true grad : {grad2(x0)}\n\tgrad: {calculate_grad2(func2,x0,delta=delta_x)},\n\t\t delta={delta_x}')
```
